```python
df.drop(columns=["RowNumber","CustomerId","Surname"],inplace=True)
```

```python
# from sklearn.preprocessing import OneHotEncoder
# ohe=OneHotEncoder()
df=pd.get_dummies(df,columns=["Geography","Gender"],drop_first=True)
```

```python
df
```

```python
# def trans(text):
#     if text=="False":
#         return 0
#     elif text=="True":
#         return 1
#     return text
        

# cols=["Geography_Germany","Geography_Spain","Gender_Male"]
# for i in cols:
#     df[i]=df[i].apply(trans)

def trans(text):
    if text == False:
        return 0
    elif text == True:
        return 1
    return text  # Return the original value if it's not "True" or "False"

cols = ["Geography_Germany", "Geography_Spain", "Gender_Male"]

# Apply the transformation
for col in cols:
    df[col] = df[col].apply(trans)
```

```python
df
```

```python
x=df.drop(columns=["Exited"])
y=df["Exited"]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
xtrain_scaled=ss.fit_transform(xtrain)
xtest_scaled=ss.transform(xtest)

```

```python
xtrain_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(11,activation="relu",input_dim=11))
model.add(Dense(11,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer="Adam",metrics=["accuracy"])
```

```python
history=model.fit(xtrain_scaled,ytrain,epochs=100,validation_split=0.2)
```

```python
y_log=model.predict(xtest_scaled)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```
